```python
import pandas as pd
d0=pd.read_csv('daily_sales_data_0.csv')
d0.head()
```

```python
# find unique products
d0['product'].unique()
```

```python
# select only pink morsel product from the dataframe
d0=d0[d0['product']=='pink morsel']
d0.info()
```

```python
d0=d0[['quantity','date','region']]
d0.groupby(['date','region']).sum()
```

```python
d1=pd.read_csv('daily_sales_data_1.csv')
```

```python
d1_pinkMorsel=d1[d1['product']=='pink morsel']
d1=d1[['quantity','date','region']]
d1
```

```python
d2=pd.read_csv('daily_sales_data_2.csv')
```

```python
d2_pinkMorsel=d2[d2['product']=='pink morsel']
d2=d2_pinkMorsel[['quantity','date','region']]
d2
```

```python
df=pd.concat([d0,d1,d2])
df.info()
```

```python
final=df.groupby(['date','region']).sum()
final['Sales']=final['quantity']*3
final=final.drop(['quantity'], axis=1)
final
```

```python
final.to_csv('response.csv')
```
